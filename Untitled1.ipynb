```julia
using Interpolations
```

```julia
f(x) = sin(x*3)
g = 0:0.1:1
v = f.(g)
```

```julia
csi = CubicSplineInterpolation(g, v; extrapolation_bc=Line())
```

```julia
csi(0.1)
```

```julia

csi([0.1, 0.2])
```

```julia
x = range(-0.1, 1.1; length=1000)
```

```julia
x
```

```julia
y = f.(x);
```

```julia
pl = plot(x, y)
scatter!(pl, g, v)
plot!(pl, x, csi(x))
```
